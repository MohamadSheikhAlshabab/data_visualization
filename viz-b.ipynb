# use car_crashes dataset:

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
car = sns.load_dataset('car_crashes')
car.head(5)
```

```python
sns.relplot(x='alcohol', y='total', hue=car.alcohol.tolist(), data=car);
plt.legend('ABCDEF', ncol=2, loc='lower right');
```

```python
car_sums = car.groupby('alcohol', as_index=False).sum()
car_mean = car.groupby('alcohol', as_index=False).mean()
car_sums
car_mean
```

```python
sns.lmplot(x='alcohol', y='total', data=car_sums);
plt.legend('ABCDEF', ncol=2, loc='upper right');
```

```python
sns.lmplot(x='alcohol', y='total', data=car_mean);
plt.legend('ABCDEF', ncol=2, loc='upper right');
```

## bar plot for car crashes dataset

```python
graph = sns.barplot(x='alcohol', y='total', data=car,ci=None);

graph.set_xticklabels(labels=list("")[1:]);
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

## bar plot for car crashes mean

```python
graph = sns.barplot(x='alcohol', y='total', data=car_mean,ci=None);

graph.set_xticklabels(labels=list("")[1:]);
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

## bar plot for car crashes sums

```python
graph = sns.barplot(x='alcohol', y='total', data=car_sums,ci=None);

graph.set_xticklabels(labels=list("")[1:]);
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

## distribution plot for car crashes 

```python
sns.distplot(car['alcohol'])
sns.distplot(car['total']);
plt.legend('ABCDEF', ncol=2, loc='upper right');
```

## joint plot for car crashes dataset:

```python
with sns.axes_style('white'):
    sns.jointplot("alcohol", "total", car, kind='kde');
    plt.legend('ABCDEF', ncol=2, loc='upper right');
```

## joint plot for car crashes dataset using kind hex:

```python
with sns.axes_style('white'):
    sns.jointplot("alcohol", "total", car, kind='hex');
    plt.legend('ABCDEF', ncol=2, loc='upper right');
```
